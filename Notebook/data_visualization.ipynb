```python
def pcaVisualize2d(x_train, s_train):
    pca = decomposition.PCA(n_components=2)
    x_train2 = x_train.reshape(x_train.shape[0], -1)
    pca.fit(x_train2)
    x_train2 = pca.transform(x_train2)
    plt.figure()
    plt.scatter(x_train2[:,0],x_train2[:,1],c = np.arange(120), marker=".")
        
def pcaVisualize3d(x_train, s_train):
    pca = decomposition.PCA(n_components=3)
    x_train2 = x_train.reshape(x_train.shape[0], -1)
    pca.fit(x_train2)
    x_train2 = pca.transform(x_train2)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection = "3d")
    ax.scatter(x_train2[:, 0], x_train2[:, 1], x_train2[:, 2], c = s_train , marker = ".")
    ax.set_xlabel("TEST1")
    ax.set_ylabel("TEST2")
    ax.set_zlabel("TEST3")
    
def tsne2d(x_train, s_train):
    x_train2 = x_train.reshape(x_train.shape[0], -1)
    X_embedded = TSNE(n_components=2).fit_transform(x_train2)
    plt.figure()
    plt.scatter(X_embedded[:,0],X_embedded[:,1],c = s_train, marker=".")
    
    
def tsne3d(x_train,s_train):
    x_train2 = x_train.reshape(x_train.shape[0], -1)
    X_embedded = TSNE(n_components=3).fit_transform(x_train2)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection = "3d")
    ax.scatter(X_embedded[:, 0], X_embedded[:, 1], X_embedded[:, 2], c = s_train , marker = ".")
    ax.set_xlabel("TEST1")
    ax.set_ylabel("TEST2")
    ax.set_zlabel("TEST3")
    
```
